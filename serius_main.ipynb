# preprocessing

In the preprocessing you do all needed steps for the dataframe to make the columns/features useable.
for example, there are some list-like features that are represented as strings, you can transform 
them into proper lists using 'literal_eval' function, for example:
here we work on all labels but since there are labels that empty or not important we doing try and expect
and we flatten all relevant labels.


```python
def preprocess2(df,process_labels=True):
    for label in df:
        try:
            df[label] = df[label].transform(literal_eval)
            df[label] = df[label].apply(lambda a: np.array(a).flatten())
        except:
            pass
    return df
```

```python
df = preprocess2(df)
df
```

# Var calculate
since we want our parameters to be matter and effective to the train we check 
if var is bigger than 11000 

```python
import statistics
for label in df:
    try:
        sample = df[label]
        var = statistics.variance(sample, xbar = -100)
        if (var > 11000 or var < 10000):
            print(label, " /-> ", var)
    except:
        pass
```
